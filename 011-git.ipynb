# Introduction of Git

https://www.petanikode.com/git-untuk-pemula/

Git adalah salah satu sistem pengontrol versi (Version Control System) pada proyek perangkat lunak yang diciptakan oleh Linus Torvalds.

Pengontrol versi bertugas mencatat setiap perubahan pada file proyek yang dikerjakan oleh banyak orang maupun sendiri.

Git dikenal juga dengan distributed revision control (VCS terdistribusi), artinya penyimpanan database Git tidak hanya berada dalam satu tempat saja.

<center><img height="400" src="image/sistem-git.png" width="650"/></center>

Semua orang yang terlibat dalam pengkodean proyek akan menyimpan database Git, sehingga akan memudahkan dalam mengelola proyek baik online maupun offline.

Dalam Git terdapat merge untuk menyebut aktifitas penggabungan kode.

Sedangkan pada VCS (Version Control System) yang terpusatâ€¦ database disimpan dalam satu tempat dan setiap perubahan disimpan ke sana.

VCS terpusat memiliki beberapa kekurangan:

Semua tim harus terkoneksi ke jaringan untuk mengakses source-code;
Tersimpan di satu tempat, nanti kalau server bermasalah bagaimana?
Karena itu, Git hadir untuk menutupi kekurangan yang dimiliki oleh VCS terpusat.

# Cara Install Git di MacOS

Buat kamu yang menggunakan MacOS, git bisa diinstal dengan `brew`.

Buka terminal lalu, ketik perintah berikut untuk menginstal Git:

```
brew install git
```

Tunggulah beberapa saat sampai proses instalasinya selesai. Setelah itu, coba ketik perintah berikut untuk mengecek versi git yang terinstal:

```
git --version
```

# Membuat Repositori

Pembuatan repositori dapat dilakukan dengan perintahÂ `git init nama-dir`. Contoh:

```
git init proyek-01
```

Perintah tersebut akan membuat direktori bernamaÂ `proyek-01`. Kalau direktorinya sudah ada, maka Git akan melakukan inisialisasi di dalam direktori tersebut.

PerintahÂ `git init`Â akan membuat sebuah direktori bernamaÂ `.git`Â di dalam proyek kita. Direktori ini digunakan Git sebagai database untuk menyimpan perubahan yang kita lakukan.

Hati-hati!

Kalau kita menghapus direktori ini, maka semua rekaman atau catatan yang dilakukan oleh Git akan hilang.

Contoh-contoh lain
------------------

Perintah berikut ini akan membuat repositori pada direktori saat iniÂ *(working directory)*.

```
git init .
```

Tanda titik (`.`) artinya kita akan membuat repository pada direktori tempat kita berada saat ini.

![Membuat repository Git](https://3.bp.blogspot.com/-IAgZktRDWBQ/WKF1raoDB4I/AAAAAAAAD3I/Uy29jWo73WARH2mV9v0rI-57tWnIlLRIACPcB/s1600/Membuat%2Brepositori%2Bgit.png)

.gitignore
----------

`.gitignore`Â merupakan sebuah file yang berisi daftar nama-nama file dan direktori yang akan diabaikan oleh Git.

Perubahan apapun yang kita lakukan terhadap file dan direktori yang sudah masuk ke dalam daftarÂ `.gitignore`Â tidak akan dicatat oleh Git.

Cara menggunakanÂ `.gitignore`, buat saja sebuah file bernamaÂ `.gitignore`Â dalam root direktori proyek/repositori.

```
/vendor/
/upload/
/cache
test.php
```

Pembuatan fileÂ `.gitignore`Â sebaiknya dilakukan di awal pembuatan repositori.

Tiga Kelompok Kondisi File dalam Git
------------------------------------

Sebelum kita membuat revisi, kita akan berkenalan dulu dengan tiga kondisi file dalam Git.

### 1\. Modified

*Modified*Â adalah kondisi di mana revisi atau perubahan sudah dilakukan, tetapi belum ditandai dan belum disimpan diÂ *version control*. Contohnya pada gambar di atas, ada tiga file HTML yang dalam kondisiÂ *modified*.

### 2\. Staged

*Staged*Â adalah kondisi di mana revisi sudah ditandai, tetapi belum disimpan diÂ *version control*. Untuk mengubah kondisi file dariÂ *modified*Â keÂ *staged*Â gunakan perintahÂ `git add nama_file`. Contoh:

```
git add index.html
```

### 3\. Committed

Committed adalah kondisi di mana revisi sudah disimpan di version control. perintah untuk mengubah kondisi file dariÂ *staged*Â keÂ *committed*Â adalahÂ `git commit`.

Mengambil Revisi dari Remote Repository
---------------------------------------

Saat kita bekerja dengan repository yang memiliki banyak kontributor, kita seharusnya mengambil dulu revisi terbaru dari repository inti agar tidak bentrok.

Misalnya begini.

Pada repository remote ada kontributor lain yang sudah menambahkan dan mengubah sesuatu di sana.

![Mengambil revisi dari repository remote](https://www.petanikode.com/img/git/remote/git-fetch-pull.png)

Maka kita harus mengambil perubahan tersebut, agar repository lokal kita tetap ter-update atau sama persis seperti repository remote.

Ada dua perintah untuk mengambil revisi dari repository remote:

1.  `git fetch [nama remote] [nama cabang]`
2.  `git pull [nama remote] [nama cabang]`

Apa perbedaannya?

PerintahÂ `git fetch`Â hanya akan mengambil revisiÂ *(commit)*Â saja dan tidak langsung melakukan penggabunganÂ *(merge)*Â terhadap repository lokal.

SedangkanÂ `git pull`Â akan mengambil revisiÂ *(commit)*Â dan langsung melakukan penggabunganÂ *(merge)*Â terhadap repository lokal.

Terus kita harus pakai yang mana?

Tergantung dari situasi dan kondisi.

Bila kita sudah membuat perubahan di repository lokal, maka sebaiknya menggunakanÂ `git fetch`Â agar perubahan yang kita lakukan tidak hilang.

Namun, bila kita tidak pernah melakukan perubahan apapun dan ingin mengambil versi terakhir dari repository remote, maka gunakanlahÂ `git pull`.
